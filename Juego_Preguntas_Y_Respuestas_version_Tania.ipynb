```python
import random

#1. Creamos un diccionario con las 23 preguntas de geografía que plantea el Instituto Cervantes en su examen de ciudadanía, utilizando las preguntas como claves y las respuestas como valores.
preguntas_respuestas = {"¿Cuál es la capital del Principado de Asturias: Oviedo, Avilés o Gijón?": "Oviedo", "¿En qué sistema montañoso está el monte Aneto: en los Pirineos, en el Sistema Central, o en Sierra Nevada?": "Los Pirineos", 
                        "En el norte de África se encuentran las ciudades autónomas de Ceuta y...: ¿Almería, Melilla o Cádiz?": "Melilla", "¿Qué tipo de clima tiene Canarias: mediterráneo, oceánico o subtropical?": "Subtropical", 
                        "¿En qué Comunidad Autónoma se encuentra la provincia de Tarragona: Galicia, Navarra o Cataluña?": "Cataluña", "¿En qué mar se encuentran las Islas Baleares: en el Cantábrico, en el Mediterráneo o en el Océano Atlántico?": "Mediterráneo", 
                        "¿Cuál es la montaña más alta de la Península Ibérica: el Mulhacén, el Aneto o el Teide": "Teide", "¿De qué Comunidad Autónoma es capital Mérida: Valencia, Cataluña o Extremadura?": "Extremadura", 
                        "¿Qué gran río desemboca en el Mar Mediterráneo: el Ebro, el Duero o el Tajo": "Ebro", "¿En qué provincia está el Parque Nacional del Monfragüe: Cáceres, Murcia o Ciudad Real?": "Cáceres",
                        "¿Dónde están Cácerez y Badajoz: en el Principado de Asturias, en Andalucía, en Extremadura?": "En Extremadura", "¿Cómo se llama la extensa llanura situada en el centro de la Península Ibérica: Marisma, Cordillera o Meseta?": "Meseta",
                        "¿En qué Comunidad Autónoma están Guadalajara y Cuenca: en Castilla León, Castilla La Mancha o Cantabria?": "Castilla La Mancha", "¿Dónde está Almería: en Andalucía, Canarias o Aragón?": "Andalucía", 
                        "Cuál es la capital de la Comunidad Autónoma de Galicia: A Coruña, Vigo o Santiago de Compostela": "Santiago de Compostela", "¿Qué clima es propio de España? Polar, Tropical o Mediterráneo": "Mediterráneo", 
                        "España se divide en comunidades y...: regiones, ciudades autónomas o distritos?": "ciudades autónomas", "De estas tres comunidades autónomas: Galicia, Aragón y La Rioja, ¿cuál de ellas es uniprovincial?": "La Rioja",
                        "¿Cuál de estos tres ríos desemboca en el Océano Atlántico: el Guadalquivir, el Manzanares o el Júcar?": "El Guadalquivir",
                        "¿Cuál de estas provincias forma parte de la Comunidad Autónoma de Castilla León: Burgos, Huesca o Guadalajara?": "Burgos",
                        "¿Las temperaturas suaves con abundantes lluvias son propias del clima...: continental, oceánico o subtropical?": "Oceánico", 
                        "¿España está entre los países de Europa más...: lluviosos, montañosos o fríos?": "Montañosos",
                        "¿Los Picos de Europa están en Andalucía, Cataluña o Asturias?": "Asturias"}
preguntas_respuestas

inicio_negrita = "\033[1m"
fin_negrita = "\033[0m"

print('¡Hola!\nTe damos la bienvenida a la ' + inicio_negrita + 'Prueba de Conocimientos Constitucionales y Socioculturales de España' + fin_negrita + ' (CCSE),\nelaborada por el ' + inicio_negrita + 'Instituto Cervantes' + fin_negrita + '. A través de este sencillo juego de Preguntas y Respuestas conocerás\nla Geografía Física y Política de \U0001F1EA\U0001F1F8')
print("_________________________________________________________________________")

nombre = input("Antes de continuar, dinos tu nombre:")
print(f"Hola, {inicio_negrita}{nombre}{fin_negrita}")
print("_________________________________________________________________________")

#3. Establecemos las condiciones necesarias para jugar.
#Definimos la PRIMERA GRAN CONDICIÓN, si quiere comenzar a jugar o no.
partida = input("Comenzamos?: s/n")
if partida.lower() == "n":
    print(f"Entendido {inicio_negrita}{nombre}{fin_negrita}, las preguntas no se van a borrar del mapa así que vuelve cuando quieras seguir\nponiendo a prueba tus conocimientos \U0001F913")
else:
    print(f"Tienes {inicio_negrita}3 intentos{fin_negrita} para responder correctamente las {inicio_negrita}10 preguntas{fin_negrita} sobre geografía española.\nProcura acentuar tus respuestas para que podamos darlas como válidas.")
    print('Una vez comenzado el juego, puedes abandonarlo pulsando la tecla x.') ## Valentina ha agregado esto
    print("_________________________________________________________________________")

    #4. Definimos la función de jugar_al_juego, usando como parámetros las claves del diccionario de preguntas y respuestas.
    #De este modo el jugador podrá repetir la partida tantas veces como quiera.
    def jugar_juego():
        intentos = 0
        aciertos = 0
        contador_preguntas = 0

        preguntas_disponibles = list(preguntas_respuestas.keys())
        #Especificamos que el juego solo realize 10 preguntas, eligíendolas al azar de las 23 totales.
        preguntas_seleccionadas = random.sample(preguntas_disponibles, 10)
    
        while contador_preguntas < 10 and intentos < 3:
            pregunta_en_juego = preguntas_seleccionadas[contador_preguntas]
            respuesta_correcta = preguntas_respuestas[pregunta_en_juego]

            respuesta_usuario = input(pregunta_en_juego + ": ")
            if respuesta_usuario == 'x':
                print('Esperamos verte pronto!') ## Valentina ha agregado esto.
                break
            #Añadimos otra condición para dar como válidas posibles respuestas del usuario como "Pirineos" o "En los Pirineos" y "Subtropical" o "Clima Subtropical".
            if respuesta_usuario.lower() == respuesta_correcta.lower() or respuesta_correcta.lower() in respuesta_usuario.lower() or respuesta_usuario.lower() in respuesta_correcta.lower():
                aciertos += 1
                contador_preguntas += 1
                print("¡Correcto!")
        
            else:
                intentos += 1
                print("La respuesta correcta era:", respuesta_correcta)
                contador_preguntas += 1

        if aciertos == 10 :
            print("=============================================================================")
            print("¡Prueba superada! ¡Has respondido correctamente todas las preguntas! \U0001F3C6")    
            print("=============================================================================")
            repetir_partida = input(f"¿Quieres repetir el pleno, {nombre}? (s/n)")
            if repetir_partida.lower() == "s" :
                print(f"Allá vamos de nuevo, {inicio_negrita}{nombre}{fin_negrita}.")
                print("_________________________________________________________________________")
                #La función se llama a sí misma y vuelva a comenzar el juego. 
                jugar_juego()
            else: 
                print(f"Perfecto {inicio_negrita}{nombre}{fin_negrita}, te esperamos para volver a recorrer España muy pronto. \U0001F44B")
                #El return hace las funciones de break y sale del juego.
                return
        elif aciertos < 10 and intentos < 3:
            print("=============================================================================")
            print("¡Prueba superada! ¡Has respondido correctamente a casi todas las preguntas! \U0001F948")  
            print("=============================================================================")
            repetir_partida = input(f"¿Quieres repetir y e intentar hacer un pleno, {nombre}? (s/n)")
            if repetir_partida.lower() == "s" :
                print(f"Alla vamos de nuevo, {inicio_negrita}{nombre}{fin_negrita}.")
                print("_________________________________________________________________________")
                #La función se llama a sí misma y vuelva a comenzar el juego. 
                jugar_juego()
            else: 
                print(f"Perfecto {nombre}, te esperamos para volver a recorrer España muy pronto. \U0001F44B")
                #El return hace las funciones de break y sale del juego.
                return
        else:
            print("=============================================================================")
            print("¡Prueba no superada! Parece que necesitas seguir recorriendo España... \U0001F622")
            print("=============================================================================")
            repetir_partida = input(f"¿Quieres volver a intentarlo, {nombre}? (s/n)")

        if repetir_partida.lower() == "s" :
            print(f"Hola de nuevo, {inicio_negrita}{nombre}{fin_negrita}")
            #La función se llama a sí misma y vuelva a comenzar el juego.  
            jugar_juego()
        else: 
            print(f"Entendido {inicio_negrita}{nombre}{fin_negrita}, las preguntas no se va a borrar del mapa así que vuelve cuando quieras seguir poniendo a prueba tus conocimientos. \U0001F913")
            #El return hace las funciones de break y sale del juego.
            return 
    
    jugar_juego()
```
